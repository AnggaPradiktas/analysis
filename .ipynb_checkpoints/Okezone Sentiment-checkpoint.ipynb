```python
train = pd.read_csv('okezone.csv', encoding= 'utf8', sep=';')
```

```python
train = train[['Title', 'Description']]
```

```python
train = train.replace(np.nan, '', regex=True)
```

```python
train['titledesc'] = train[['Title', 'Description']].apply(lambda x: ' '.join(x), axis = 1)
```

```python
train
```

```python
tags_lower = train['titledesc'].str.lower().str.cat(sep=' ')
```

```python
tags_remove_pun = re.sub('[^A-Za-z]+', ' ', tags_lower)
```

```python
stop_words = list(get_stop_words('id'))
```

```python
nltk_words = list(stopwords.words('indonesian'))
```

```python
stop_words.extend(nltk_words)
```

```python
word_tokens_tags = word_tokenize(tags_remove_pun)
```

```python
filtered_sentence_tags = [w_tags for w_tags in word_tokens_tags if not w_tags in stop_words]
```

```python
filtered_sentence_tags = []
```

```python
for w_tags in word_tokens_tags:
    if w_tags not in stop_words:
        filtered_sentence_tags.append(w_tags)
```

```python
without_double_char = [word_tags for word_tags in filtered_sentence_tags if len(word_tags) > 2]
```

```python
cleaned_data = [word_tags for word_tags in without_double_char if not word_tags.isnumeric()]
```

```python
top_N = 10000
```

```python
word_dist_tags = nltk.FreqDist(cleaned_data)
```

```python
result_tags = pd.DataFrame(word_dist_tags.most_common(top_N), columns=['Word', 'Frequency'])
```

```python
plt.figure(figsize=(18,10))
sns.set_style('darkgrid')
plt.title('Okezone Topics')

ax = sns.barplot(x = 'Word', y = 'Frequency', data=result_tags.head(10))

rects = ax.patches

for rect in rects:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    
    space = 5
    
    va = 'bottom'
    
    label = '{:.1f}'.format(y_value)
    
    plt.annotate(
        label,
        (x_value, y_value),
        xytext=(0, space),
        textcoords='offset points',
        ha = 'center',
        va = va)
```

```python
bloblist_desc = list()

train_descr_str=train['Title'].astype(str)
for row in train_descr_str:
    blob = TextBlob(row)
    an = blob.translate(from_lang='id', to='en')
    bloblist_desc.append((row,an.sentiment.polarity, an.sentiment.subjectivity))
    train_polarity_desc = pd.DataFrame(bloblist_desc, columns = ['sentence','sentiment','polarity'])
```

```python
 
def f(train_polarity_desc):
    if train_polarity_desc['sentiment'] > 0:
        val = "Positive"
    elif train_polarity_desc['sentiment'] == 0:
        val = "Neutral"
    else:
        val = "Negative"
    return val

train_polarity_desc['Sentiment_Type'] = train_polarity_desc.apply(f, axis=1)

plt.figure(figsize=(5,10))
sns.set_style("whitegrid")
plt.title('Okezone Sentiment')
ax = sns.countplot(x="Sentiment_Type", data=train_polarity_desc, palette=["C2", "k", "C0"])
```
