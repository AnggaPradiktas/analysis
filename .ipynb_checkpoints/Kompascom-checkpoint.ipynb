```python
#Multiple Page Scraping Iterations
```

```python
pages = [str(i) for i in range(1,37)]
```

```python
newstitles = []
dates = []
```

```python
start_time = time()
requests = 0
```

```python
for page in pages:
    # Make a get request
    response = get('https://indeks.kompas.com/tag/pupr/desc/' + page)
    
    sleep(randint(8,15))
    
    requests += 1
    elapsed_time = time() - start_time
    print('Request: {}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)
    
    soup1 = BeautifulSoup(response.text, 'html.parser')
    
    news_container = soup1.find_all('div', class_='article__list clearfix')
    
    for news in news_container:
        if news.find('div', class_='article__list__title') is not None:
            newstitle = news.h3.text
            newstitles.append(newstitle)
            
            date = news.find_all('div', class_='article__date')
            dates.append(date)
```

```python
a = {"newstitle":newstitles, "date":dates}



kompasnews = pd.DataFrame.from_dict(a, orient='index')

kompasnews.transpose()
```

```python
kompasnews = kompasnews.transpose()
```

```python
kompasnews.insert(1, "media", "Kompascom")
```

```python
kompasnews = kompasnews.drop_duplicates(subset=['newstitle'], keep='first')
```

```python
import sklearn
from sklearn.model_selection import train_test_split
train, test = train_test_split(kompasnews, test_size=0.33, random_state=42)
```

```python
kompasnews.drop(columns =["date"], inplace = True) 
```

```python
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace(',', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('.', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('!', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('?', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace(':', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace(';', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('Rp', '')
```

```python
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('di', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('Ini', '')
kompasnews['newstitle'] = kompasnews['newstitle'].str.replace('dan', '')
```

```python
kompasnews_text = [text for text in kompasnews[kompasnews['media'] == 'Kompascom']
['newstitle']]
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
joint_words = ' '.join(kompasnews_text)

separated_words = word_tokenize(joint_words)
```

```python
kompasnews_counts = Counter(separated_words)
```

```python
kompasnews_common_words = [word[0] for word in kompasnews_counts.most_common(10)]
kompasnews_common_counts = [word[1] for word in kompasnews_counts.most_common(10)]
```

```python
fig = plt.figure(figsize=(18,10))
sns.barplot(x=kompasnews_common_words, y=kompasnews_common_counts)
plt.title('kompascom')
plt.show()
```
