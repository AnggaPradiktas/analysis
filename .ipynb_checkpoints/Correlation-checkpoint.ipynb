```python
df1 = df[['videoCategoryId','durationSec','viewCount','likeCount','dislikeCount','commentCount']]
```

```python
df1.corr()
```

```python
h_labels = [x.replace('_', ' ').title() for x in 
            list(df1.select_dtypes(include=['number', 'bool']).columns.values)]

fig, ax = plt.subplots(figsize=(10,6))
_ = sns.heatmap(df1.corr(), annot=True, xticklabels=h_labels, yticklabels=h_labels, cmap=sns.cubehelix_palette(as_cmap=True), ax=ax)
```

```python
fig, ax = plt.subplots()
_ = plt.scatter(x=df['viewCount'], y=df['likeCount'], color=PLOT_COLORS[2], edgecolors="#000000", linewidths=0.3)
_ = ax.set(xlabel="Views", ylabel="Likes")
```

```python
cat_df_gb = df['day_of_week'].value_counts().reset_index()
plt.figure(figsize=(10,10))
sns.set_style("darkgrid")
ax = sns.barplot(x=cat_df_gb['index'],y=cat_df_gb['day_of_week'], data=cat_df_gb)
plt.ylabel("Number of Videos")
plt.xlabel("Days")
plt.title("Number of Videos posted Each Days")

rects = ax.patches

for rect in rects:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    
    space = 5
    
    va = 'bottom'
    
    label = '{:.1f}'.format(y_value)
    
    plt.annotate(
        label,
        (x_value, y_value),
        xytext=(0, space),
        textcoords='offset points',
        ha = 'center',
        va = va)

```

```python
groupedvalues=df.groupby('day_of_week').sum().reset_index()

plt.figure(figsize=(10,10))
sns.set_style("darkgrid")
ax1 = sns.barplot(y='viewCount', x='day_of_week', data=groupedvalues)
plt.xlabel("Days")
plt.ylabel("Views")
plt.title("Number of Views Each Days")



rects = ax1.patches

for rect in rects:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    
    space = 5
    
    va = 'bottom'
    
    label = '{:.1f}'.format(y_value)
    
    plt.annotate(
        label,
        (x_value, y_value),
        xytext=(0, space),
        textcoords='offset points',
        ha = 'center',
        va = va)
```

```python
df['likes_log'] = np.log(df['likeCount'] + 1)
df['views_log'] = np.log(df['viewCount'] + 1)
df['dislikes_log'] = np.log(df['dislikeCount'] + 1)
df['comment_log'] = np.log(df['commentCount'] + 1)

plt.figure(figsize = (12,6))

plt.subplot(221)
g1 = sns.distplot(df['views_log'])
g1.set_title("VIEWS LOG DISTRIBUITION", fontsize=16)

plt.subplot(224)
g2 = sns.distplot(df['likes_log'],color='green')
g2.set_title('LIKES LOG DISTRIBUITION', fontsize=16)

plt.subplot(223)
g3 = sns.distplot(df['dislikes_log'], color='r')
g3.set_title("DISLIKES LOG DISTRIBUITION", fontsize=16)

plt.subplot(222)
g4 = sns.distplot(df['comment_log'])
g4.set_title("COMMENTS LOG DISTRIBUITION", fontsize=16)

plt.subplots_adjust(wspace = 0.2, hspace = 1,top = 1)

plt.show()
```
